Q.9 What is the Difference Between a Primary Key and a Foreign Key?


ANS.9 DIFFERENCE BETWEEN PRIMARY KEY AND FOREIGN KEY

      PRIMARY KEY

      IT IS UNIQUELY IDENTIFIES EACH RECORD IN A TABLE.
      MUST HAVE UNIQUE VALUES . CAN NOT HAVE NULL VALUES .
      ITS PURPOSWE IS TO IDENTIFIES RECODS .LACATION EXISTS IN SAME TABLE . ONLY ONE PEIMARY KEY ALLOWED IN PER TABLE .

      FOREIGN KEY

      IT LINKS ONE TABLE TO ANOTHER . CAN HAVE UPLICATE VALUES,CAN HAVE NULL VALUES . ITS PURPOSE IS MAINTAIN RELATIONSHIPS BETWEEN TABLES . LOCATED IN REFERENCES ANOTHER TABLES PRIMARY KEY . MULTIPLE FOREIGN KEYS ALLOWED.
